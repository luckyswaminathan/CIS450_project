```python
import re
import json
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')

stop_words = list(set(stopwords.words('english') +
                 stopwords.words('spanish') +
                 stopwords.words('french') +
                 stopwords.words('german')))
```

```python
def process_tweet(tweet):
    tweet_without_urls = re.sub(r'https?:\/\/\S+', '', tweet)

    hashtags = re.findall(r'(#\w+)', tweet_without_urls)
    retweets = re.findall(r'(https://\S+)', tweet)
    ats = re.findall(r'(@\w+)', tweet_without_urls)
    keywords = re.findall(r'\b(\w+)\b', tweet_without_urls)

    keywords = [word.lower() for word in keywords
                if word.lower() not in [tag.strip('#').lower() for tag in hashtags]
                and word.lower() not in [at.strip('@').lower() for at in ats]
                and word.lower() not in stop_words]

    return hashtags, retweets, ats, keywords
```

```python
final_df.dropna(subset=['tweet'], inplace=True)
final_df.dropna(subset=['user_id'], inplace=True)
final_df.dropna(subset=['topic'], inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df['processed'] = final_df['tweet'].apply(process_tweet)
```

```python
documents = []
for _, row in final_df.iterrows():
    doc = {
        '_id': row['tweet_id'],
        'hashtags': row['processed'][0] if row['processed'][0] else None,
        'links': row['processed'][1] if row['processed'][1] else None,
        'ats': row['processed'][2] if row['processed'][2] else None,
        'keywords': row['processed'][3] if row['processed'][3] else None
    }
    doc = {k: v for k, v in doc.items() if v is not None}
    documents.append(doc)

import json
json_output = json.dumps(documents, indent=2)
print(json_output)
```
